```python
import pandas as pd
import numpy as np

#importando banco de dado sobre reprovados.
Quatitativo_reprovacao = pd.read_excel('quatitativo_reprovacao_PRO.xlt',sep=';', encoding='latin-1')
```

```python
#retirando valores indesejados.
Quatitativo_reprovacao = Quatitativo_reprovacao[Quatitativo_reprovacao['ANO'] != 2020]
Quatitativo_reprovacao = Quatitativo_reprovacao[Quatitativo_reprovacao['MATRICULAS'] != 0]
Quatitativo_reprovacao = Quatitativo_reprovacao[Quatitativo_reprovacao['MATRICULAS'] != 1]
#Quatitativo_reprovacao = Quatitativo_reprovacao[Quatitativo_reprovacao['ANO'] > 2003]
```

```python
#Gerando colunas necessárias para a graficação.
Quatitativo_reprovacao['TAXA_DE_APROVACAO'] = round(Quatitativo_reprovacao['APROVACOES']/ Quatitativo_reprovacao['MATRICULAS'],2)
Quatitativo_reprovacao['TAXA_REPROVACAO_POR_NOTA'] = round(Quatitativo_reprovacao['REPROVACOES NOTA']/ Quatitativo_reprovacao['MATRICULAS'],2)
Quatitativo_reprovacao['TAXA_REPROVACAO_POR_FALTA'] = round((Quatitativo_reprovacao['REPROVACOES NOTA FALTA']+Quatitativo_reprovacao['REPROVACOES FALTA'])/ Quatitativo_reprovacao['MATRICULAS'],2)
Quatitativo_reprovacao['TAXA_DE_TRANCAMENTO'] = round(Quatitativo_reprovacao['TRANCAMENTOS']/ Quatitativo_reprovacao['MATRICULAS'],2)
Quatitativo_reprovacao['ANO/SEMESTRE'] = Quatitativo_reprovacao.apply(lambda Quatitativo_reprovacao:'%s.%s' % (Quatitativo_reprovacao['ANO'],Quatitativo_reprovacao['SEMESTRE']),axis=1)

Quatitativo_reprovacao1 = Quatitativo_reprovacao.groupby(['ANO/SEMESTRE', 'DISCIPLINA','DESCRICAO']).TAXA_DE_APROVACAO.mean()
Quatitativo_reprovacao1 = pd.DataFrame(Quatitativo_reprovacao1)
Quatitativo_reprovacao1 = Quatitativo_reprovacao1.reset_index()

Quatitativo_reprovacao1['contador'] = 1
selecao_materias = Quatitativo_reprovacao1.groupby('DESCRICAO').contador.sum()
selecao_materias = pd.DataFrame(selecao_materias)
selecao_materias = selecao_materias.reset_index()
selecao_materias = selecao_materias[selecao_materias.contador >= 5]
selecao_materias.tail()

result = pd.merge(Quatitativo_reprovacao1, selecao_materias, left_on='DESCRICAO', right_on ='DESCRICAO', how='inner')
result = result.drop(columns= ['contador_x'])
result = result.rename(columns={"contador_y": "Frequência da matéria"})
result.tail()
```

```python
#Gerando colunas necessárias para a graficação.
Quatitativo_reprovacao2 = Quatitativo_reprovacao
Quatitativo_reprovacao2 = Quatitativo_reprovacao2.groupby(['ANO/SEMESTRE', 'DISCIPLINA','DESCRICAO']).TAXA_REPROVACAO_POR_NOTA.mean()
Quatitativo_reprovacao2 = pd.DataFrame(Quatitativo_reprovacao2)
Quatitativo_reprovacao2 = Quatitativo_reprovacao2.reset_index()

Quatitativo_reprovacao3 = Quatitativo_reprovacao
Quatitativo_reprovacao3 = Quatitativo_reprovacao3.groupby(['ANO/SEMESTRE', 'DISCIPLINA','DESCRICAO']).TAXA_REPROVACAO_POR_FALTA.mean()
Quatitativo_reprovacao3 = pd.DataFrame(Quatitativo_reprovacao3)
Quatitativo_reprovacao3 = Quatitativo_reprovacao3.reset_index()

Quatitativo_reprovacao4 = Quatitativo_reprovacao
Quatitativo_reprovacao4 = Quatitativo_reprovacao4.groupby(['ANO/SEMESTRE', 'DISCIPLINA','DESCRICAO']).TAXA_DE_TRANCAMENTO.mean()
Quatitativo_reprovacao4 = pd.DataFrame(Quatitativo_reprovacao4)
Quatitativo_reprovacao4 = Quatitativo_reprovacao4.reset_index()
```
